# Chapter 1 - Understanding large language models

## 1 What is an LLM?

An LLM is a neural network designed to understand, generate, and respond to human-like text.

"Large" refers to both the model's size in terms of parameters and the immense dataset on which it's trained.

LLMs utilize an architecture called the *transformer*, which allows them to pay selective attention to different parts of the input when making predictions.

## 2 Applications of LLMs

+ Machine translation  
+ Generation of novel texts  
+ Sentiment analysis  
+ Text summarization  
+ Content creation: writing fiction, articles, computer code  
+ Power sophisticated chatbots and virtual assistants  
+ Effective knowledge retrieval

LLMs are in short invaluable for automating almost any task that involves parsing and generating text.

## 3 Stages of building and using LLMs

1. Pre-training: Train on a large corpus of text data ("raw", unlabeled text)
This stage creates an initial pre-trained LLM, often called a *base* or *foundation model*.

2. Fine tuning: train the LLM on labeled data
+ *Instruction fine-tuning*: the labeled dataset consists of instruction and answer pairs
+ *Classification fine-tuning*: the labeled dataset consists of texts and associated class labels

## 4 Introducting the transformer architecture

2017 paper "Attention Is All You Need"

The transformer architecture consists of two submodules: an encoder and a decoder.
+ Encoder: processes the input text and encodes it into a series of numerical representations of vectors that capture the contextual information of the input
+ Decoder: takes these encoded vectors and generates the output text.

Both the encoder and decoder consist of many layers connected to a so-called self-attention mechanism.

The self-attention mechanism allows the model to weigh the importance of different words or tokens in a sequence relative to each other.

Later variants of the transformer architecture, such as BERT (*bidirectional encoder representations from transformers*) and the various GPT models (*generative pretrained transformers*) build on this concept to adapt this architecture for different tasks.

Zero-shot learning: ability to generalize to completely unseen tasks without any prior specific examples.
Few-shot learning: involves learning from a minimal number of examples the user provides as input.

## 6 A closer look at the GPT architecture

The general GPT architecture os just the decoder part without the encoder.
Since decoder-style models like GPT generate text by predicting text one word at a time, they are considered a type of *autoregressive* model.
Autoregressive models incorporate their previous outputs as inputs for future predictions.
Consequentl, in GPT, each new word is chosen based on the sequence that precedes it, which improves the coherence of the resulting text.
Architectures such as GPT-3 are significantly larger than the original transformer model. For instance, the original transformer repeated the encoder and decoder blocks six times. GPT-3 has 96 transformer layers and 175 billion parameters in total.
The ability to perform tasks that the model wasn't explicitly trained to perform is called an *emergent behavior* (e.g. GPT models are capable of performing translation tasks).

## 7 Building a large language model

Three main stages:
1. Implementing the LLM architecture and data preparation process
2. Pretraining an LLM to create a foundational model
3. Fine-tuning the foundation model to become a personal assistant or a text classifier
